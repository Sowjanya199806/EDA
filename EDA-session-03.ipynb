$Take$

- Take is used to extract the particular index of data

- take has 1 parameter called axis

- axis=1 represents columns

- axis=0 represent rows

- by default axis=0 is there

- In below example we give 100,200,300 and we did not provide axis

- it understood we are talking about rows

```python
visa_df.take([100,200,300])
```

```python
visa_df.take([100,200,300],axis=1)
```

```python
visa_df.take([2,3,5,7],axis=1)
```

```python
visa_df.take([100:105])  
# drawback
# we cant take rows & columns at a time 
# with the single syntax not possible we have to take 2 takes
```

```python
visa_df.take([100,200,300]).take([2,5,7],axis=1)
```
