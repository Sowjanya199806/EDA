```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
visa_df=pd.read_csv(r"C:\Users\sowja\Downloads\Visadataset - Visadataset (2).csv")
visa_df
```

```python
ed=visa_df['education_of_employee']=='High School'
visa_df[ed]
```

```python
unique_ed=visa_df['education_of_employee'].unique()
for i in unique_ed:
    ed=visa_df['education_of_employee']==i
    val=len(visa_df[ed])
    print(f"{i}  : {val}")
```

```python
unique_ed=visa_df['education_of_employee'].unique()
count_list=[]
for i in unique_ed:
    edu=visa_df['education_of_employee']==i
    val=len(visa_df[edu])
    count_list.append(val)

#####################################################
education_df=pd.DataFrame(zip(unique_ed,count_list),
                         columns=['education_of_employee','Count'])
######################################################
education_df.to_csv('education_of_employee_df.csv',index=False)
```
